```python
import bs4 as bs
import urllib.request 
```

```python
sauce=urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup=bs.BeautifulSoup(sauce,'lxml')
```

```python
table=soup.table
```

```python
print(table)
```

```python
lisst=[]
table_rows=table.find_all('tr')
for tr in table_rows:
    td=tr.find_all('td')
    row=[i.text for i in td]
    lisst.append(row)
    print(row)
```

```python
print(lisst)
```

```python
df = pd.DataFrame(lisst,columns=['Postal Code', 'Borough', 'Neighborhood'])
```

```python
df.head()
```

```python
df = df.replace('\n','', regex=True)
```

```python
df.head()
```

```python
df=df[df.Borough != 'Not assigned']
```

```python
df.head()
```

```python
df=df.drop(df.index[0])
```

```python
df.head()
```

```python
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df=df.groupby(['Postal Code']) ##Groupby using Postal code
```
